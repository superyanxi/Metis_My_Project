```python
#Data Cleaning and Preparation
#1.Standardizing the type of values
#2.Filling missing data
#3.Merging tables
import pandas as pd
import numpy as np
import csv
from sklearn.utils import shuffle
```

```python
#First table: Details of all coupons
coupon_info = pd.read_csv('coupon_list_train.csv')
coupon_info.head()
```

```python
#Transferring from non-numeric value to numeric value as Dummy varibles
coupon_info["CAPSULE_TEXT"] = coupon_info["CAPSULE_TEXT"].map({
    'グルメ':int(1),'ヘアサロン':int(2),
    'エステ':int(3),'リラクゼーション':int(4),
    '宅配':int(5),'レジャー':int(6),
    'ホテル':int(7),'旅館':int(8),
    'ネイル・アイ':int(8),'通学レッスン':int(9),
    'その他':int(10),'ロッジ':int(11),
    'WEBサービス':int(12),'貸別荘':int(13),
    'ギフトカード':int(14)})
coupon_info["CAPSULE_TEXT"].fillna(int(15),inplace=True)
```

```python
coupon_info["ken_name"] = coupon_info["ken_name"].map({
    '大阪府':int(1),'東京都':int(2),
    '神奈川県':int(3),'兵庫県':int(4),
    '愛知県':int(5),'埼玉県':int(6),
    '山梨県':int(7),'広島県':int(8),
    '三重県':int(9),'宮城県':int(10),
    '群馬県':int(11),'京都府':int(12),
    '長崎県':int(13),'北海道':int(14),
    '千葉県':int(15),'高知県':int(16),
    '大分県':int(17),'静岡県':int(18),
    '沖縄県':int(19),'福島県':int(20),
    '福岡県':int(21),'福井県':int(22),
    '秋田県':int(23),'茨城県':int(24),
    '長野県':int(25),'山形県':int(26),
    '熊本県':int(27),'岡山県':int(28),
    '和歌山県':int(29),'岐阜県':int(30),
    '滋賀県':int(31),'岩手県':int(32),
    '岩手県':int(33),'奈良県':int(34),
    '石川県':int(35)})
```

```python
coupon_info0 = coupon_info[['COUPON_ID_hash','CAPSULE_TEXT','PRICE_RATE',
                            'CATALOG_PRICE','DISCOUNT_PRICE',
                            'DISPPERIOD','USABLE_DATE_THU','USABLE_DATE_FRI',
                           'USABLE_DATE_SAT','USABLE_DATE_SUN',
                           'USABLE_DATE_HOLIDAY','USABLE_DATE_BEFORE_HOLIDAY',
                           'ken_name',]]
```

```python
coupon_info0.head()
```

```python
#Dealing with the missing values
null_columns=coupon_info0.columns[coupon_info0.isnull().any()]
coupon_info0[null_columns].isnull().sum()
```

```python
coupon_info0 = coupon_info0.dropna(subset=['ken_name'])
```

```python
coupon_info0 = coupon_info0.fillna(0)
null_columns=coupon_info0.columns[coupon_info0.isnull().any()]
coupon_info0[null_columns].isnull().sum()
```

```python
#Table 2: Purchase History
purchase_data = pd.read_csv('coupon_detail_train.csv')
purchase_data = purchase_data[['USER_ID_hash','COUPON_ID_hash']]
purchase_data = purchase_data.rename(index=str,columns={'COUPON_ID_hash':'COUPON_ID_hash_purchase'})
purchase_data = purchase_data.drop_duplicates()
purchase_data.head()
```

```python
purchase_data.info()
```

```python
#Table 3: Browse History
browse_hist = pd.read_csv('coupon_visit_train.csv')
browse_hist = browse_hist[['USER_ID_hash','VIEW_COUPON_ID_hash']]
browse_hist = browse_hist.drop_duplicates()
pd.DataFrame.info(browse_hist)
```

```python
browse_hist.head()
```

```python
#Table 4: Customer Profile
customer = pd.read_csv('user_list.csv')
customer = customer[['USER_ID_hash','SEX_ID','AGE']]
```

```python
#Left Join table of browse history with table of Purchase History
join0 = pd.merge(browse_hist,purchase_data,left_on = ['USER_ID_hash','VIEW_COUPON_ID_hash'],
         right_on = ['USER_ID_hash','COUPON_ID_hash_purchase'],how = 'left')
```

```python
join0['COUPON_ID_hash_purchase'] = join0['COUPON_ID_hash_purchase'].fillna(0)
```

```python
for index,value in join0.iterrows():
    if value['COUPON_ID_hash_purchase'] !=0:
        join0.iloc[index]['COUPON_ID_hash_purchase'] = 1
```

```python
#Problem of Imbalance of classes. Undersampling is needed to obatain a subset of the orginal dataframe
join0.groupby(['COUPON_ID_hash_purchase']).count()
```

```python
temp = pd.DataFrame()
rows = np.random.choice
temp = join0.groupby(['COUPON_ID_hash_purchase']).get_group(0)
temp = temp.sample(129995)
temp2 = join0.groupby(['COUPON_ID_hash_purchase']).get_group(1)
join1 = pd.concat([temp,temp2])
join1 = join1.reindex()
join1.groupby(['COUPON_ID_hash_purchase']).count()
```

```python
#Joining the tables of information of coupon and customer
join2 = pd.merge(join1,coupon_info0,left_on='VIEW_COUPON_ID_hash',right_on = 'COUPON_ID_hash',how = 'left')
join2 = join2.dropna()
join2 = join2.drop_duplicates(subset = ['USER_ID_hash','VIEW_COUPON_ID_hash'])
```

```python
join3 = pd.merge(join2,customer,left_on ='USER_ID_hash',right_on ='USER_ID_hash', how = 'left')
join3 = join3.dropna()
join3 = join3[['USER_ID_hash', 'VIEW_COUPON_ID_hash','CAPSULE_TEXT', 'PRICE_RATE', 'CATALOG_PRICE',
       'DISCOUNT_PRICE', 'DISPPERIOD', 'USABLE_DATE_THU', 'USABLE_DATE_FRI',
       'USABLE_DATE_SAT', 'USABLE_DATE_SUN', 'USABLE_DATE_HOLIDAY',
       'USABLE_DATE_BEFORE_HOLIDAY', 'ken_name', 'SEX_ID', 'AGE','COUPON_ID_hash_purchase']]
join3['COUPON_ID_hash_purchase'] = join3['COUPON_ID_hash_purchase'].astype(int)
join3['SEX_ID'] = join3['SEX_ID'].astype(int)
join3 = shuffle(join3)
```
